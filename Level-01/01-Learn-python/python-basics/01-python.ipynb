```
# guessing number challenge

import random
jackpot = random.randint(1,100)

def guessing_number(count = 1) :
  number = int(input("enter number :"))
  if(number > jackpot ):
    print("guess The number lower")
    return guessing_number(count + 1)
  elif(number< jackpot):
    print("The number is greater")
    return guessing_number(count + 1)
  else :
    print("correct number guessed")
    return count


noOfAttempts = guessing_number()
print("atempt to guess correct number : ",noOfAttempts)
```
