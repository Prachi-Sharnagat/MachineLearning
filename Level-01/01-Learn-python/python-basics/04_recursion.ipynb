```python
#iterative way :
def multiply(a,b):
  result = 0
  for i in range(b): #run 0 to b-1
    result = result + a
  return result

multiply(4,5)
```

```python
# using recursion
def mutiply(a,b):
  if(b==0):
    return 0;
  else :
    return a + mutiply(a,b-1)

mutiply(4,5)
```

```python
def fact(number):
  if(number==1):
    return 1;
  else:
    return number*fact(number-1)
fact(5)
```

```python
# plaindrome by recursion
def isPalindrome(str):
  n = len(str)
  i = 0
  while(i<n/2):
      if(str[i]!=str[n-i-1]):
        return 0;
      else:
        i = i + 1;
  return 1

str = 'madam'

isPalindrome(str)
```

```python
#rabit que
import time
def rabitSolution(month):

  if(month == 1 or month==0):
    return 1;
  else:
    sum = rabitSolution(month-2) + rabitSolution(month-1);
    return sum
start = time.time()
print(rabitSolution(4))
end = time.time()
# calculate timetaken by the given code
print(end-start)
# inefficient code : taken long time for big data
```

```python
import time
def fib(m,d):
  if m in d :
    return d[m]
  else:
    d[m] = fib(m-1,d) + fib(m-2,d);
    return d[m]

d = {0:1, 1:1}
start = time.time()
print(fib(45,d))
end = time.time()
print(end - start)
print(d)
```
