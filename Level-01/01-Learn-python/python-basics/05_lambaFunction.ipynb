```python
# higher order function


# even number ssum
# odd sum
# sum of x % 3 == divisible
def allSum(l):
  even_sum = 0
  odd_sum = 0
  div3_sum = 0
  for i in l :
    if i%2==0:
      even_sum = even_sum + i
    else :
      odd_sum = odd_sum + i

  print(even_sum)
  print(odd_sum)

  for j in l:
    if(j%3 == 0) :
     div3_sum = div3_sum + j;
  print(div3_sum)

l = [11,3,72,71,56]
print(allSum(l))
```

```python
# use lambda
def return_sum(func,l):
  result = 0
  for i in l:
    if func(i): # it means if funct have i then true display then add it in result
      result = result + i
  return result
l = [2,3,4,53,1]
x = lambda x : x%2==0
y = lambda y : y%2!=0
z = lambda z : z%3 == 0

print(return_sum(x,l))
print(return_sum(y,l))
print(return_sum(z,l))
```
